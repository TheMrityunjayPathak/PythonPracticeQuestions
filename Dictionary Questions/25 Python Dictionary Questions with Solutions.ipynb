**Question 1:**
Create an empty dictionary.

**Solution:**
```python
my_dict = {}
```

**Question 2:**
Create a dictionary with key-value pairs for a person's name and age.

**Solution:**
```python
person = {"name": "John", "age": 30}
```

**Question 3:**
Access the value associated with the key "age" in the `person` dictionary.

**Solution:**
```python
age = person["age"]
```

**Question 4:**
Check if a key exists in the `person` dictionary.

**Solution:**
```python
if "name" in person:
    print("Key 'name' exists.")
```

**Question 5:**
Add a new key-value pair "city" with the value "New York" to the `person` dictionary.

**Solution:**
```python
person["city"] = "New York"
```

**Question 6:**
Remove the key-value pair with the key "age" from the `person` dictionary.

**Solution:**
```python
del person["age"]
```

**Question 7:**
Iterate through the keys in the `person` dictionary.

**Solution:**
```python
for key in person:
    print(key)
```

**Question 8:**
Iterate through the values in the `person` dictionary.

**Solution:**
```python
for value in person.values():
    print(value)
```

**Question 9:**
Iterate through both keys and values in the `person` dictionary.

**Solution:**
```python
for key, value in person.items():
    print(key, value)
```

**Question 10:**
Create a dictionary with default values using the `dict.fromkeys()` method.

**Solution:**
```python
defaults = dict.fromkeys(["name", "age", "city"], "Unknown")
```

**Question 11:**
Get the value associated with a key using the `dict.get()` method with a default value.

**Solution:**
```python
age = person.get("age", 0)
```

**Question 12:**
Count the number of occurrences of each character in a string using a dictionary.

**Solution:**
```python
text = "hello"
char_count = {}
for char in text:
    char_count[char] = char_count.get(char, 0) + 1
```

**Question 13:**
Merge two dictionaries.

**Solution:**
```python
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}
merged_dict = {**dict1, **dict2}
```

**Question 14:**

**Solution:**
