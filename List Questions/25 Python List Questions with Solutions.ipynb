1. **Question:** Given a list `nums`, find the sum of all elements in the list.
```python
def sum_of_elements(nums):
    return sum(nums)
```

2. **Question:** Write a function to find the maximum element in a list.
```python
def max_element(nums):
    return max(nums)
```

3. **Question:** Remove all duplicates from a list.
```python
def remove_duplicates(nums):
    return list(set(nums))
```

4. **Question:** Count the occurrences of an element in a list.
```python
def count_occurrences(nums, target):
    return nums.count(target)
```

5. **Question:** Reverse a list in-place (modify the original list).
```python
def reverse_list_in_place(nums):
    nums.reverse()
```

6. **Question:** Check if a list is empty.
```python
def is_list_empty(nums):
    return len(nums) == 0
```

7. **Question:** Find the index of the first occurrence of an element in a list.
```python
def find_first_index(nums, target):
    return nums.index(target)
```

8. **Question:** Create a new list containing only the even numbers from the original list.
```python
def filter_even_numbers(nums):
    return [num for num in nums if num % 2 == 0]
```

9. **Question:** Create a new list with the squares of each element in the original list.
```python
def square_elements(nums):
    return [num ** 2 for num in nums]
```

10. **Question:** Sort a list in ascending order.
```python
def sort_list(nums):
    return sorted(nums)
```

11. **Question:** Check if all elements in the list are positive.
```python
def all_positive(nums):
    return all(num > 0 for num in nums)
```

12. **Question:** Find the second-largest element in a list.
```python
def second_largest(nums):
    return sorted(nums)[-2]
```

13. **Question:** Remove all occurrences of a specific element from the list.
```python
def remove_all_occurrences(nums, target):
    return [num for num in nums if num != target]
```

14. **Question:** Find the common elements between two lists.
```python
def find_common_elements(list1, list2):
    return list(set(list1) & set(list2))
```

15. **Question:** Check if a list is a palindrome (reads the same forward and backward).
```python
def is_palindrome(nums):
    return nums == nums[::-1]
```

16. **Question:** Flatten a list of lists into a single list.
```python
def flatten_list(lists):
    return [item for sublist in lists for item in sublist]
```

17. **Question:** Calculate the product of all elements in the list.
```python
from functools import reduce
import operator
def product_of_elements(nums):
    return reduce(operator.mul, nums, 1)
```

18. **Question:** Remove the first and last elements from a list.
```python
def remove_first_and_last(nums):
    return nums[1:-1]
```

19. **Question:** Find the difference between two lists.
```python
def list_difference(list1, list2):
    return list(set(list1) - set(list2))
```

20. **Question:** Merge two lists and sort them in ascending order.
```python
def merge_and_sort(list1, list2):
    return sorted(list1 + list2)
```

21. **Question:** Find the longest sublist in a list of lists.
```python
def longest_sublist(lists):
    return max(lists, key=len)
```

22. **Question:** Split a list into chunks of a given size.
```python
def split_into_chunks(nums, chunk_size):
    return [nums[i:i+chunk_size] for i in range(0, len(nums), chunk_size)]
```

23. **Question:** Count the number of odd and even numbers in a list.
```python
def count_odd_even(nums):
    odd_count = len([num for num in nums if num % 2 != 0])
    even_count = len(nums) - odd_count
    return odd_count, even_count
```

24. **Question:** Find the index of the last occurrence of an element in a list.
```python
def find_last_index(nums, target):
    return len(nums) - 1 - nums[::-1].index(target)
```

25. **Question:** Rotate a list to the left by a given number of positions.
```python
def rotate_left(nums, positions):
    positions %= len(nums)
    return nums[positions:] + nums[:positions]
```
