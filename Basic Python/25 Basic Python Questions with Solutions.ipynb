
4. **Question**: Calculate the factorial of a given number.
```python
# Explanation:
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```


5. **Question**: Find the largest among three numbers entered by the user.


8. **Question**: Write a Python program to check if a given string is a palindrome.
```python
# Explanation:
def is_palindrome(s):
    return s == s[::-1]
string = input("Enter a string: ")
if is_palindrome(string):
    print("Palindrome")
else:
    print("Not a palindrome")
```


9. **Question**: Count the number of vowels in a given string.
```python
# Explanation:
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for char in s if char in vowels)
string = input("Enter a string: ")
print("Number of vowels:", count_vowels(string))
```


12. **Question**: Check if a given number is prime.
```python
# Explanation:
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
num = int(input("Enter a number: "))
if is_prime(num):
    print("Prime")
else:
    print("Not prime")
```


18. **Question**: Check if two strings are anagrams.


22. **Question**: Check if a given number is a perfect square.
```python
# Explanation:
def is_perfect_square(num):
    return int(num**0.5)**2 == num
num = int(input("Enter a number: "))
if is_perfect_square(num):
    print("Perfect square")
else:
    print("Not a perfect square")
```


25. **Question**: Print the Fibonacci sequence up to a given number of terms.
```python
# Explanation:
def fibonacci(n):
    fib_seq = [0, 1]
    while len(fib_seq) < n:
        fib_seq.append(fib_seq[-1] + fib_seq[-2])
    return fib_seq
num_terms = int(input("Enter the number of terms: "))
fibonacci_seq = fibonacci(num_terms)
print("Fibonacci sequence:", fibonacci_seq)
```
