1. Write a Python program to reverse a string.
```python
def reverse_string(input_string):
    return input_string[::-1]
```

2. Write a Python program to check if a string is a palindrome.
```python
def is_palindrome(input_string):
    return input_string == input_string[::-1]
```

3. Write a Python program to count the number of vowels in a string.
```python
def count_vowels(input_string):
    vowels = "aeiouAEIOU"
    return sum(1 for char in input_string if char in vowels)
```

4. Write a Python program to find the most frequent character in a string.
```python
def most_frequent_char(input_string):
    char_count = {}
    for char in input_string:
        char_count[char] = char_count.get(char, 0) + 1
    return max(char_count, key=char_count.get)
```

5. Write a Python program to remove all duplicates from a string and return the result.
```python
def remove_duplicates(input_string):
    return "".join(sorted(set(input_string), key=input_string.index))
```

6. Write a Python program to check if a string contains only digits.
```python
def contains_only_digits(input_string):
    return input_string.isdigit()
```

7. Write a Python program to count the occurrences of each word in a sentence.
```python
def count_word_occurrences(input_string):
    words = input_string.split()
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
```

8. Write a Python program to capitalize the first letter of each word in a sentence.
```python
def capitalize_first_letter(input_string):
    return " ".join(word.capitalize() for word in input_string.split())
```

9. Write a Python program to find the longest word in a sentence.
```python
def longest_word(input_string):
    words = input_string.split()
    return max(words, key=len)
```

10. Write a Python program to check if two strings are anagrams.
```python
def are_anagrams(str1, str2):
    return sorted(str1) == sorted(str2)
```

11. Write a Python program to count the number of occurrences of a specific substring in a string.
```python
def count_substring_occurrences(input_string, substring):
    count = 0
    index = 0
    while True:
        index = input_string.find(substring, index)
        if index == -1:
            break
        count += 1
        index += 1
    return count
```

12. Write a Python program to find the index of the first occurrence of a substring in a string.
```python
def find_substring(input_string, substring):
    return input_string.find(substring)
```

13. Write a Python program to replace all occurrences of a substring with another substring.
```python
def replace_substring(input_string, old_substring, new_substring):
    return input_string.replace(old_substring, new_substring)
```

14. Write a Python program to find the length of the last word in a sentence.
```python
def length_of_last_word(input_string):
    words = input_string.split()
    if words:
        return len(words[-1])
    return 0
```

15. Write a Python program to reverse the order of words in a sentence.
```python
def reverse_words_order(input_string):
    words = input_string.split()
    return " ".join(reversed(words))
```

16. Write a Python program to check if a string is a valid email address.
```python
import re
def is_valid_email(email):
    pattern = r'^[\w\.-]+@[a-zA-Z\d\.-]+\.[a-zA-Z]{2,}$'
    return bool(re.match(pattern, email))
```

17. Write a Python program to check if a string is a valid URL.
```python
import re
def is_valid_url(url):
    pattern = r'^https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+'
    return bool(re.match(pattern, url))
```

18. Write a Python program to find the first non-repeated character in a string.
```python
from collections import Counter
def first_non_repeated_char(input_string):
    char_count = Counter(input_string)
    for char in input_string:
        if char_count[char] == 1:
            return char
    return None
```

19. Write a Python program to remove all leading and trailing whitespaces from a string.
```python
def trim_whitespace(input_string):
    return input_string.strip()
```

20. Write a Python program to find the common characters between two strings.
```python
def common_characters(str1, str2):
    return "".join(char for char in set(str1) if char in str2)
```

21. Write a Python program to find the second most frequent character in a string.
```python
def second_most_frequent_char(input_string):
    char_count = {}
    for char in input_string:
        char_count[char] = char_count.get(char, 0) + 1
    sorted_char_count = sorted(char_count.items(), key=lambda x: x[1], reverse=True)
    return sorted_char_count[1][0] if len(sorted_char_count) >= 2 else None
```

22. Write a Python program to check if a string contains only unique characters.
```python
def has_unique_characters(input_string):
    return len(set(input_string)) == len(input_string)
```

23. Write a Python program to find the longest common prefix among a list of strings.
```python
def longest_common_prefix(strings):
    if not strings:
        return ""
    prefix = strings[0]
    for string in strings[1:]:
        i = 0
        while i < len(prefix) and i < len(string) and prefix[i] == string[i]:
            i += 1
        prefix = prefix[:i]
    return prefix
```

24. Write a Python program to check if a string is a valid IPv4 address.
```python
import socket
def is_valid_ipv4(ipv4):
    try:
        socket.inet_pton(socket.AF_INET, ipv4)
        return True
    except socket.error:
        return False
```

25. Write a Python program to find the first non-repeated character in a string using OrderedDict.
```python
from collections import OrderedDict
def first_non_repeated_char_ordered(input_string):
    char_count = OrderedDict()
    for char in input_string:
        char_count[char] = char_count.get(char, 0) + 1
    for char, count in char_count.items():
        if count == 1:
            return char
    return None
```
